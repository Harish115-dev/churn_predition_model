```python
df_temp=df.copy()
```

```python
df_temp
```

```python
tenur=df_temp["tenure"]
```

```python
plt.hist(tenur)
plt.xlabel("tenur months")
plt.ylabel("counts")
plt.title("Histogram of tenur")

```

```python
churn=df_temp["Churn"]
count=churn.value_counts()
```

```python


plt.bar(count.index, count.values)
plt.xlabel("churn")
plt.ylabel("count")
plt.title("count of churn")
```

```python
sns.countplot(data=df_temp, x='Contract', hue='Churn')
plt.title("Contract Type vs Churn")
plt.xlabel("Contract Type")
plt.ylabel("Churn Count")
```

```python

sns.stripplot(x="Churn", y="MonthlyCharges", data=df_temp, jitter=True)
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")
plt.title("Monthly Charges by Churn")
plt.show()
```

```python
sns.countplot(data=df_temp, x='Partner', hue='Churn')
plt.title("partner vs Churn")
plt.xlabel("partner")
plt.ylabel("Churn Count")
```

```python
sns.countplot(data=df_temp, x='Dependents', hue='Churn')
plt.title("Dependents vs Churn")
plt.xlabel("dependents")
plt.ylabel("Churn Count")
```

```python
sns.countplot(data=df_temp, x='gender', hue='Churn')
plt.title("gender vs Churn")
plt.xlabel("gender")
plt.ylabel("Churn Count")
```
